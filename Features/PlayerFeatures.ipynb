```python
# IDENTIFYING PLAYERS WHO WON GAMES. 
# SINCE THIS DATA ALSO CONTAINS EACH MATCH WHERE THE PLAYER PLAYED, ALL "WINNING_PLAYER" VALUES PER MATCH PER TEAM WILL BE THE SAME 

playerApps_Outcomes_df.loc[playerApps_Outcomes_df["result"] == "home team win", "match_winning_team"] = playerApps_Outcomes_df["home_team_name"]
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["result"] == "away team win", "match_winning_team"] = playerApps_Outcomes_df["away_team_name"]
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["result"] == "draw", "match_winning_team"] = "Draw"
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["team_name"] == playerApps_Outcomes_df["match_winning_team"], "game_outcome"] = "win"
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["team_name"] != playerApps_Outcomes_df["match_winning_team"], "game_outcome"] = "loss"

# ADDING A SORT TO THE GROUP STAGES FOR ORGANIZATION

playerApps_Outcomes_df.loc[playerApps_Outcomes_df["stage_name"] == "group stage", "stage_name_sort"] = 1
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["stage_name"] == "second group stage", "stage_name_sort"] = 1.5
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["stage_name"] == "round of 16", "stage_name_sort"] = 2
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["stage_name"] == "quarter-finals", "stage_name_sort"] = 3
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["stage_name"] == "semi-finals", "stage_name_sort"] = 4
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["stage_name"] == "final", "stage_name_sort"] = 5
playerApps_Outcomes_df.loc[playerApps_Outcomes_df["stage_name"] == "third-place match", "stage_name_sort"] = 4.5

```

```python
# GROUPED TEAMS BY NAME, MATCH YEAR, STAGE OF WC, AND GAME OUTCOME, WITH NUMBER OF MATCHES PLAYED
grouped_by_stage_df = pd.DataFrame(playerApps_Outcomes_df[["team_name", "stage_name", "match_id", "game_outcome", "match_year","stage_name_sort", ]].groupby(["team_name","match_year","stage_name", "game_outcome"]).nunique("match_id").sort_values(["team_name","match_year","stage_name_sort"], ascending=True))

historic_wins_df = pd.DataFrame(playerApps_Outcomes_df[["team_name", "stage_name", "match_id", "game_outcome","stage_name_sort"]].groupby("team_name")["game_outcome"].apply(lambda x: (x=="win").count()))
historic_losses_df = pd.DataFrame(playerApps_Outcomes_df[["team_name", "stage_name", "match_id", "game_outcome","stage_name_sort"]].groupby("team_name")["game_outcome"].apply(lambda x: (x=="loss").count()))
historic_wins_losses =historic_losses_df.merge(historic_wins_df,left_on="team_name", right_on="team_name", how="inner")
historic_wins_losses.rename(columns={"game_outcome_x":"losses", "game_outcome_y":"win"})
historic_wins_losses.head(40)
```

#### 20 players with World Cup Appearances since 1970 are associated with multiple teams. 

This largely stems from the end of the Soviet Union in 1991 and the resulting destabilization and shifting political boundaries in the Balkan region. 

Thriteen of these players played for West Germany prior to the unification of Germany in Aug 1990 (after that year's World Cup), afterwhich, all represented Germany in the World Cups after 1990. Similarly, two players played for the Soviet Union in 1990, and represented Russia in 1994. 
    
Four players who played for Yugoslavia in the 1990s went on to play for different countries during the 20 years in which:
The Socialist Federal Republic of Yugoslavia was dissolved and re-founded as the Federal Republic of Yugoslavia (1992). The Federal Republic of Yugoslavia was transformed into the State Union of Serbia and Montenegro (2003) which existed until Montenegro seceded (2006), creating two independent nations. Oh, and also during this time, the Bosnian Wars broke out, during which Croatia's borders with Yugoslavia / Serbia and Montenegro shifted. One player played for Serbia and Montenegro in 2006 and then for Serbia in 2010
