```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
#Read in csv files
matches_features_df= pd.read_csv('Data/bookings.csv')
goal_features_df= pd.read_csv('Data/goals.csv')
```

```python
# Filter matches_features dataframe to have match date >= 1970
matches_features_df['match_date'] = pd.to_datetime(matches_features_df['match_date'])
```

```python
matches_features_df['year'] = matches_features_df['match_date'].dt.year
```

```python
matches_features_df = matches_features_df[matches_features_df['year']>=1970]
```

```python
#Get the shape of matches df
matches_features_df.shape
```

```python
#Filter goal_features dataframe to have match date >= 1970
goal_features_df['match_date'] = pd.to_datetime(goal_features_df['match_date'])
```

```python
goal_features_df['year'] = goal_features_df['match_date'].dt.year
```

```python
goal_features_df = goal_features_df[goal_features_df['year']>=1970]
```

```python
# In match_features_df, create average number of cards per game by match per country
```

```python
matches_features_df.head()
```

```python
# Group the matches and get the average of the yellow cards
x = matches_features_df.groupby(['match_name'])['yellow_card'].mean()
```

```python
df = pd.DataFrame(x)

df = df.reset_index()
df.columns = ['Match_Name','Average Yellow_Card']
```

```python
df
```

```python
# In goal_features_df, likelihood a country will score early using minute_regulation and match_period
```

```python
#Likelihood of scoring goal was taken by mean of the minute regulation
y = goal_features_df.groupby(['match_name','match_period'])['minute_regulation'].mean()
```

```python
df1 = pd.DataFrame(y)

df1 = df1.reset_index()
df1.columns = ['Match_Name','match_period','likelihood']
```

```python
df_merged = pd.merge(df,df1,on='Match_Name')
```
