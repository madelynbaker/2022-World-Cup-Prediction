#### 20 players with World Cup Appearances since 1970 are associated with multiple teams. 

This largely stems from the end of the Soviet Union in 1991 and the resulting destabilization and shifting political boundaries in the Balkan region. 

Thriteen of these players played for West Germany prior to the unification of Germany in Aug 1990 (after that year's World Cup), afterwhich, all represented Germany in the World Cups after 1990. Similarly, two players played for the Soviet Union in 1990, and represented Russia in 1994. 
    
Four players who played for Yugoslavia in the 1990s went on to play for different countries during the 20 years in which:
The Socialist Federal Republic of Yugoslavia was dissolved and re-founded as the Federal Republic of Yugoslavia (1992). The Federal Republic of Yugoslavia was transformed into the State Union of Serbia and Montenegro (2003) which existed until Montenegro seceded (2006), creating two independent nations. Oh, and also during this time, the Bosnian Wars broke out, during which Croatia's borders with Yugoslavia / Serbia and Montenegro shifted. One player played for Serbia and Montenegro in 2006 and then for Serbia in 2010
