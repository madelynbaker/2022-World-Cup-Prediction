```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
#Read in csv files
matches_features_df= pd.read_csv('Data/bookings.csv')
goal_features_df= pd.read_csv('Data/goals.csv')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Filter matches_features dataframe to have match date >= 1970
matches_features_df['match_date'] = pd.to_datetime(matches_features_df['match_date'])
```

```python
matches_features_df['year'] = matches_features_df['match_date'].dt.year
```

```python
matches_features_df = matches_features_df[matches_features_df['year']>=1970]
```

```python
#Get the shape of matches df
matches_features_df.shape
```

```python
#Filter goal_features dataframe to have match date >= 1970
goal_features_df['match_date'] = pd.to_datetime(goal_features_df['match_date'])
```

```python
goal_features_df['year'] = goal_features_df['match_date'].dt.year
```

```python
goal_features_df = goal_features_df[goal_features_df['year']>=1970]
```

```python
# In match_features_df, create average number of cards per game by match per country
```

```python
matches_features_df
```

```python
#Group the matches and get the average of the yellow cards 
#Divide matches into team_name and assign yellow card to team who recv'd yellow card; get the count and average
```

```python
df_divide_match = matches_features_df.groupby('team_name')['yellow_card'].agg(['count','mean']).reset_index()
```

```python
df_divide_match.head()
```

```python
df_divide_match.rename(columns={"team_name": "Team_Name", "count": "Yellow_Card_Count", "mean": "Avg_Yellow_Card"},inplace=True)
```

```python
#Print new dataframe
df_divide_match
```

```python
# In goal_features_df, likelihood a violation would be commited
```

```python
#Likelihood of violation was taken by mean of the minute regulation
y = goal_features_df.groupby('team_name')['minute_regulation'].agg(['count','mean'])
```

```python
df1 = pd.DataFrame(y)

df1 = df1.reset_index()
df1.columns = ['Team_Name','Count_Violation','Likelihood_Violation']
```

```python
df1
```

```python
goal_features_df
```

```python
# Combine the dataframes together 
```

```python
df_merged = pd.merge(df_divide_match,df1,on='Team_Name')
```
