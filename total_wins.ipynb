```python
# run dependencies

import pandas as pd
from sqlalchemy import create_engine
```

```python
historic_csv = "Data/matches.csv"
```

```python
history = pd.read_csv(historic_csv)
history
```

```python
# iterating the columns
for col in history.columns:
    print(col)
```

```python
history['match_date'] = pd.to_datetime(history['match_date'], format='%Y-%m-%d')
```

```python
history_df =  history.loc[(history["match_date"] >= '1970-01-01')]
history_df
```

```python
history_df['tournament_id'].str.contains('WC')
```

```python
home_wins = history_df[["home_team_name", "home_team_win"]].groupby("home_team_name").count()
home_wins
```

```python
away_wins = history_df[["away_team_name", "away_team_win"]].groupby("away_team_name").count()
```

```python
home_wins.index.names = ['Country']
home_wins
```

```python
away_wins = away_wins.rename(columns = {'away_team_win':'Wins'})
```

```python
away_wins.index.names = ['Country']
away_wins
```

```python
# away_wins = away_wins.rename(columns = {'away_team_name': 'Country'})
```

```python
country_wins = pd.merge(home_wins, away_wins, on='Country',how='inner')
```

```python
country_wins
```

```python
country_wins = country_wins['home_team_win'] + country_wins['Wins']
```

```python
country_wins
```

```python
total_wins = pd.DataFrame(country_wins)
```
