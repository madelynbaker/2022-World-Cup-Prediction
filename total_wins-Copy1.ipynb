# Making dataframes based on WIN margins

```python
home_scores = history_df[['home_team_name', 'home_team_score_margin', 'home_team_win']]
home_scores = home_scores[home_scores.home_team_win != 0]
home_scores
```

```python
# HOME TEAM MARGIN WINS
home_avg_win_by = home_scores[["home_team_name", "home_team_score_margin"]].groupby("home_team_name").mean()
```

```python
home_avg_win_by = home_avg_win_by.reset_index()
home_avg_win_by = home_avg_win_by.rename(columns = {'home_team_name': 'Country','home_team_score_margin': 'win_margin'})
home_avg_win_by
```

```python
away_scores = history_df[['away_team_name', 'away_team_score_margin', 'away_team_win']]
away_scores = away_scores[away_scores.away_team_win != 0]
away_scores
```

```python
# AWAY TEAM MARGIN WINS
away_avg_win_by = away_scores[["away_team_name", "away_team_score_margin"]].groupby("away_team_name").mean()
away_avg_win_by
```

```python
away_avg_win_by = away_avg_win_by.reset_index()
away_avg_win_by = away_avg_win_by.rename(columns = {'away_team_name': 'Country','away_team_score_margin': 'win_margin_away'})
```

```python
total_win_by = pd.merge(home_avg_win_by, away_avg_win_by, on = 'Country')
total_win_by['win_margins'] = total_win_by['win_margin']+total_win_by['win_margin_away']
total_win_by = total_win_by[['Country', 'win_margins']]
total_win_by
```

# Making dataframes based on LOSE margins

```python
home_losses = history_df[['home_team_name', 'home_team_score_margin', 'home_team_win']]
home_losses = home_losses[home_losses.home_team_win != 1]
```

```python
# HOME TEAM MARGIN LOSSES
home_avg_lose_by = home_losses[["home_team_name", "home_team_score_margin"]].groupby("home_team_name").mean()
```

```python
home_avg_lose_by = home_avg_lose_by.reset_index()
home_avg_lose_by = home_avg_lose_by.rename(columns = {'home_team_name': 'Country','home_team_score_margin': 'lose_margin_home'})
```

```python
away_losses = history_df[['away_team_name', 'away_team_score_margin', 'away_team_win']]
away_losses = away_losses[away_losses.away_team_win != 1]
```

```python
# AWAY TEAM MARGIN LOSSES
away_avg_lose_by = away_losses[["away_team_name", "away_team_score_margin"]].groupby("away_team_name").mean()
```

```python
away_avg_lose_by = away_avg_lose_by.reset_index()
away_avg_lose_by = away_avg_lose_by.rename(columns = {'away_team_name': 'Country','away_team_score_margin': 'loss_margin_away'})
```

```python
total_lose_by = pd.merge(home_avg_lose_by, away_avg_lose_by, on = 'Country')
total_lose_by['lose_margins'] = total_lose_by['lose_margin_home']+total_lose_by['loss_margin_away']
total_lose_by = total_lose_by[['Country', 'lose_margins']]
total_lose_by
```

# merging margin wins and losses !!!!!

```python
total_score_margins = pd.merge(total_win_by, total_lose_by, on = 'Country', how = 'inner')
total_score_margins = total_score_margins[(total_score_margins['Country'].isin(teams_2022))]
total_score_margins
```

```python
adv1 = history_df.groupby(["stage_name","home_team_name"]).count()
adv1 = adv1.reset_index()
adv1 = adv1.rename(columns = {'home_team_name':'country'})
adv1
```

```python
adv1 = adv1[['stage_name','country','match_id']]
adv1
```

```python
adv2 = history_df.groupby(["stage_name","away_team_name"]).count()
adv2 = adv2.reset_index()
adv2 = adv2.rename(columns = {'away_team_name':'country'})
adv2
```

```python
adv2 = adv2[['stage_name','country','match_id']]
adv2
```

```python
stage_count = pd.merge(adv1, adv2, on=['stage_name','country'],how='outer')
stage_count = stage_count.fillna(0)
```

```python
stage_count['adv_counts'] = stage_count['match_id_x']+stage_count['match_id_y']
stage_count
```

```python
adv_df = stage_count[['stage_name','country','adv_counts']]
adv_df
```
